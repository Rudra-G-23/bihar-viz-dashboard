# Gender 

```python
fig = px.pie(
    fdf['Gender_label'].value_counts(),
    values='count',
    names='Gender_label',
    title="Gender Pie Chart"
)

fig.update_layout(showlegend=True)

fig.show()
```

# yr of edu

```python
year_of_edu_df  = fdf['Years_of_Education'].value_counts()

year_of_edu_df
```

```python
fig = px.bar(
    year_of_edu_df,
    x=year_of_edu_df['Years_of_Education'],
    y=year_of_edu_df['count'],
    range_y=[0, 8_000]
)

fig.show()
```

```python
fig = px.box(
    fdf,
    x='Used_Internet_Last_30_Days',
    y='Years_of_Education',
    color='Gender',
    title="Last used Internet w/ Gender w/ Age"
)

fig.update_layout(showlegend=True)

fig.update_xaxes(
    tickvals=[1, 2],
    ticktext=['Yes', 'No']
) 

fig.show()
```

# Marital_Status

```python
fig = px.pie(
    fdf['Marital_Status_label'].value_counts(),
    values='count',
    names='Marital_Status_label',
    title="Marital Status Pie Chart"
)

fig.update_layout(showlegend=True)

fig.show()
```

```python
fig = px.density_heatmap(
    fdf[['Marital_Status', 'Education_Level']], 
    x='Marital_Status', 
    y='Education_Level', 
    text_auto=True, 
    title='Marital Status by Education Level'
)

fig.update_xaxes(
    tickvals=[1, 2, 3, 4],
    ticktext=['Never Married', 'Currently Married', 'Widowed', 'Divorced/Separated']
)

fig.show()
```
